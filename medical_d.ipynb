```python
diseases = ['Fungal infection']
disease_count = {}  # empty dictionary to store counts

for disease in diseases:
    if disease in disease_count:
        disease_count[disease] += 1  # increase count if already exists
    else:
        disease_count[disease] = 1   # start count if new

print(disease_count)
```
